## Handling Missing Values
Dataset: Building permits issued in San Francisco

```python
# importing required modules
import numpy as np
import pandas as pd
np.random.seed(0)
```

```python
# reading data
df = pd.read_csv("Datasets/Building_Permits.csv")
df.head(10)
```

```python
# getting no of missing values and displaying for first 10 columns
missing = df.isnull().sum()
missing
```

```python
# percentage of missing values
total_cells = np.product(df.shape)
total_missing = missing.sum()
percent = (total_missing/total_cells)*100
print(percent)
```

```python
# dropping rows that contain null values
df.dropna()
```

```python
# droping columns that have null values
new_data = df.dropna(axis=1)
new_data
```

```python
print("Columns of original dataset: ", df.shape[1])
print("Columns of new dataset: ", new_data.shape[1])
```

```python
subset = df.loc[:, ["Street Number", "Unit", "Proposed Units", "Number of Existing Stories", "Estimated Cost", "Revised Cost", "Number of Proposed Stories", "Supervisor District"]]
subset.head()
```

```python
# fill null values with 0
subset.fillna(0)
```

```python
# fill null values with previous values else with 0
subset.fillna(method='bfill', axis=0).fillna(0)
```
