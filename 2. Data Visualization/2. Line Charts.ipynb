## Line Charts
Dataset: Global daily streams Spotify

```python
# importing required modules
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading the data
df = pd.read_csv("Datasets/spotify.csv", index_col="Date", parse_dates=True)
```

```python
# printing first 10 rows of data
df.head(10)
```

```python
# printing last 10 rows of data
df.tail(10)
```

```python
# plotting line graph for data which shows daily global streams for each song
sns.lineplot(data=df)
```

```python
# setting the width and height of the figure or graph
plt.figure(figsize=(14,6))

# giving a title fo rthe figure
plt.title("Line graph for global streams of each song")

# plotting line graph for data
sns.lineplot(data=df)
```

```python
# printing all columns
list(df.columns)
```

```python
# plotting line graph only for two songs

# setting the width and height of figure
plt.figure(figsize=(14,6))

# line graph for global streams of 'Shape of you'
sns.lineplot(data=df["Shape of You"], label="Shape of You")

# line graph for global streams of 'Despacito'
sns.lineplot(data=df["Despacito"], label="Despacito")

# label for horizontal axis
plt.xlabel("Date")
```
