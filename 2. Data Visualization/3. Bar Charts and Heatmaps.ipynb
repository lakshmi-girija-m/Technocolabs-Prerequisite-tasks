## Bar charts and Heatmaps
Dataset: Flight delays

```python
# importing required modules
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# reading the data
df = pd.read_csv("Datasets/flight_delays.csv", index_col="Month")

# printing first 10 rows of data
df.head(10)
```

```python
# setting the width and height of figure
plt.figure(figsize=(10,6))

# title of bar chart
plt.title("Average arrival delay of Spirits Airlines flights for each month")

# plotting bar graph for Spirits Airlines with months as horizontal axis and bars showing the average delay time
sns.barplot(x=df.index, y=df["NK"])

# label for y axis
plt.ylabel("Arrival delay (in minutes)")
```

```python
# setting the width and height of figure
plt.figure(figsize=(14,7))

# title of heatmap
plt.title("Average arrival delay of Spirits Airlines flights for each month")

# displaying heatmap with average delay time for all airlines
sns.heatmap(data=df, annot=True)

# label for x axis
plt.xlabel("Airlines")
```
