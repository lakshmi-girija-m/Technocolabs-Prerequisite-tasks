Dataset: Tips

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
# distribution of total_bill column
sns.distplot(tips['total_bill'], kde=True, hist=True)
```

```python
# tip given by different sex
sns.barplot(x='sex', y='tip', data=tips)
```

```python
# tip given by smoker and non smoker
sns.barplot(x='smoker', y='tip', data=tips)
```

```python
# tips given according to day of week
sns.barplot(x='day', y='tip', data=tips)
```

```python
# tips given based on meal time
sns.barplot(x='time', y='tip', data=tips)
```

```python
# tips based on size
sns.barplot(x='size', y='tip', data=tips)
```

```python
# tips fo rsex and smoker or not
sns.barplot(x='sex', y='tip', data=tips, hue='smoker')
```

```python
# tip for total_bill and sex
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex')
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time')
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style="time")
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day')
```

```python
sns.barplot(data=tips, x='day', y='tip', hue='sex')
```
